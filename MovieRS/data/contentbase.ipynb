```python
movies_data['tags'] = movies_data['overview'] + movies_data['genre']
```

```python
count = movies_data["title"].count()
```

```python
movies_data=movies_data.reset_index()
```

```python
def remove_stopword(text):
    lemmat = nltk.stem.WordNetLemmatizer()
    pre_text = []
    words = text.split()
    for word in words:
        if word not in list_stopwords:
            pre_text.append(word)
        text2 = ' '.join(lemmat.lemmatize(x) for x in pre_text)
    return text2
```

```python
for i in range(count):
    movies_data.loc[i,"tags"] = remove_stopword(movies_data["tags"][i])
```
